```python
import keras
import numpy as np
import matplotlib.pyplot as pp
```

```python
x_shape=(500)
X=np.random.random(x_shape)
#평균
mean=np.random.randint(-100,100)
#표준정규
std=700
X=np.random.random(x_shape)

X=np.random.normal(mean,std,x_shape)


pp.plot(np.arange(0,500,1),X,"ro")
```

```python
def random():
    return np.random.randint(-1000,1000)*np.random.random(1)
a=random()
b=random()
c=random()
d=random()
e=random()
def f(x):
    return 4*((1/16)*x**3)+2*((1/7)*x**4)
Y=f(X)
print(a,b)
pp.plot(X,Y,"ro")
```

```python
model=keras.models.Sequential()
model.add(layer=keras.layers.Dense(100,input_shape=(1,)))
model.add(layer=keras.layers.Dense(100,activation='relu'))
model.add(layer=keras.layers.Dense(200,activation='relu'))
model.add(layer=keras.layers.Dense(1000,activation='relu'))
model.add(layer=keras.layers.Dense(1))
model.compile(loss=keras.losses.binary_crossentropy,optimizer='adam')
```

```python
model.fit(X,Y,epochs=20000)
```

```python
Y_=model.predict(X)
pp.plot(X,Y_,"ro")
```
